2. Write a function that will plot the ranking of a given roller coaster over time as a line. Your function should take a roller coaster's name and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with `"El Toro"` as the roller coaster name and the wood ranking DataFrame. What issue do you notice? Update your function with an additional argument to alleviate the problem, and retest your function.

```python
# 2
# Create a function to plot rankings over time for 1 roller coaster
def ranking_plot(name,park_name,rank_df):
    coaster_rankings = rank_df[(rank_df['Name'] == name) & (rank_df['Park'] == park_name)]
    fig,ax = plt.subplots(figsize = (10,10))
    ax.plot(coaster_rankings['Year of Rank'], coaster_rankings['Rank'])
    ax.invert_yaxis()
    ax.set_yticks([1,2,3])
    plt.xticks(size = 12)
    plt.yticks(size = 12)
    plt.xlabel('Year', size = 14)
    plt.ylabel('Rank', size = 14)
    plt.title(f"{name} Roller Coaster Rank by Year", size = 16)
    plt.show()
# Create a plot of El Toro ranking over time
ranking_plot("El Toro",'Six Flags Great Adventure',wood)
```

Most notably, we inverted the scale on the y-axis as the lower value rank is better. We also limited the x ticks to integer values as decimals are irrelevant in this case. Furthermore, the title is formed of an f-strings with the roller coaster argument being a variable in it.

```python
# 3
# Create a function to plot rankings over time for 2 roller coasters
def ranking_plots(coaster_1, coaster_2, park_1, park_2, rank_df):
    coaster1_ranking = rank_df[(rank_df['Name'] == coaster_1) & (rank_df['Park'] == park_1)]
    coaster2_ranking = rank_df[(rank_df['Name'] == coaster_2) & (rank_df['Park'] == park_2)]
    fig,ax = plt.subplots(figsize = (8,8))
    ax.plot(coaster1_ranking['Year of Rank'], coaster1_ranking['Rank'], label = f"{coaster_1} Roller Coaster")
    ax.plot(coaster2_ranking['Year of Rank'], coaster2_ranking['Rank'], label = f"{coaster_2} Roller Coaster")
    ax.set_yticks([1,2,3,4])
    plt.xticks(size = 12)
    plt.yticks(size = 12)
    ax.invert_yaxis()
    plt.xlabel('Year', size = 14)
    plt.ylabel('Rank', size = 14)
    plt.title(f"Rankings of {coaster_1} & {coaster_2} Roller Coasters over Time", size = 16)
    plt.legend(loc = 'best')
# Create a plot of El Toro and Boulder Dash roller coasters
ranking_plots("El Toro","Boulder Dash","Six Flags Great Adventure","Lake Compounce",wood)
```

4. Write a function that will plot the ranking of the top `n` ranked roller coasters over time as lines. Your function should take a number `n` and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   For example, if `n == 5`, your function should plot a line for each roller coaster that has a rank of `5` or lower.
   
   Call your function with a value of `n` and either the wood ranking or steel ranking DataFrame.

```python
# 4
# Create a function to plot top n rankings over time
def topN_rankings(n, rank_df):
    topn = rank_df[rank_df['Rank'] <= n]
    fig,ax = plt.subplots(figsize = (10,10))
    for coaster in set(topn['Name']):
        coaster_ranking = topn[topn['Name'] == coaster]
        ax.plot(coaster_ranking['Year of Rank'], coaster_ranking['Rank'], label = coaster)
        ax.invert_yaxis()
        ax.set_yticks([1,2,3,4,5])
        plt.xlabel('Year of Rank', size = 12)
        plt.ylabel('Rank', size = 12)
        plt.title(f"Ranking of Top {n} Roller Coasters Over Time", size = 14)
        plt.legend(loc=4,fontsize = 8)
# Create a plot of top n rankings over time
topN_rankings(5,wood)
```

```python
# 6
# Create a function to plot histogram of column values
def column_histogram(df, col):
    plt.hist(df[col])
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Roller Coaster {col}')
    plt.show()
    plt.clf()
# Create histogram of roller coaster speed
column_histogram(roller_coasters,'speed')
# Create histogram of roller coaster length
column_histogram(roller_coasters,'length')
# Create histogram of roller coaster number of inversions
column_histogram(roller_coasters,'num_inversions')
# Create a function to plot histogram of height values
def height_histogram(df):
    heights = df[df['height'] <= 140]['height'].dropna()
    plt.hist(heights)
    
# Create a histogram of roller coaster height
height_histogram(roller_coasters)
plt.ylabel('Frequency')
plt.xlabel('Height')
plt.title('Histogram of Roller Coaster Height')
plt.show()
plt.clf()
```

```python
# 8
# Create a function to plot a pie chart of status.operating
def pie_status(df):
    df = df[(df['status'] == "status.operating") | (df['status'] == "status.closed.definitely")]
    plt.figure(figsize = (8,6))
    proportions = df['status'].value_counts(normalize = True)
    plt.pie(proportions, labels = ['Operating','Closed Definitely'], autopct = '%.2f',
           textprops= {'fontsize':12})
    plt.title('Pie Chart of Roller Coaster Operating Status', size = 16)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
# Create pie chart of roller coasters
pie_status(roller_coasters)
```

9. `.scatter()` is another useful function in matplotlib that you might not have seen before. `.scatter()` produces a scatter plot, which is similar to `.plot()` in that it plots points on a figure. `.scatter()`, however, does not connect the points with a line. This allows you to analyze the relationship between two variables. Find [`.scatter()`'s documentation here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).

   Write a function that creates a scatter plot of two numeric columns of the roller coaster DataFrame. Your function should take the roller coaster DataFrame and two-column names as arguments. Make sure to include informative labels that describe your visualization.
   
   Call your function with the roller coaster DataFrame and two-column names.

```python
# 9
# Create a function to plot scatter of any two columns
def scatter_coasters(df,column1,column2):
    plt.figure(figsize = (8,6))
    plt.scatter(df[column1],df[column2])
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.title(f"Scatter Plot of {column2} vs {column1}")
# Create a function to plot scatter of speed vs height
def speed_height_scatter(df,speed,height):
    plt.figure(figsize = (8,6))
    df['height'] = df['height'][df['height'] < 140]
    plt.scatter(df[height],df[speed])
    plt.xlabel('Height', size = 14)
    plt.ylabel('Speed', size = 14)
    plt.xticks(size = 12)
    plt.yticks(size = 12)
    plt.title('Scatter Plot of Roller Coaster Speed vs Height', size = 16)
# Create a scatter plot of roller coaster height by speed
speed_height_scatter(roller_coasters,'height','speed')
```

#### Question 2: Do different seating types result in higher/longer/faster roller coasters?

To answer this question, we can group the data by different seating types and find the height, length, and speed of roller coasters.

```python
mean_grouped = roller_coasters.groupby('seating_type')[['height','speed','length']].mean().reset_index()
mean_grouped = mean_grouped[mean_grouped.seating_type != "na"]
```

```python
print(mean_grouped)
```

```python
?plt.bar
```

```python
plt.figure(figsize = (10,10))
plt.bar(x=mean_grouped.seating_type, height=mean_grouped.height)
plt.xticks(rotation = 45, size = 11)
plt.yticks(size = 12)
plt.ylabel('Roller Coaster Height', size = 12)
plt.title('Bar Plot of Roller Coaster Height by Seating Type', size = 16)
plt.show()
```

From the plot above, we can see that different seating types are associated with different roller coaster heights. There are specific seating types which lead to the highest heights of roller coasters. These include Wing, Floorless, and 4th Dimension.

```python
plt.figure(figsize = (10,10))
plt.bar(x=mean_grouped.seating_type, height=mean_grouped.speed)
plt.xticks(rotation = 45, size = 11)
plt.yticks(size = 12)
plt.ylabel('Roller Coaster Speed', size = 12)
plt.title('Bar Plot of Roller Coaster Speed by Seating Type', size = 16)
plt.show()
```

Similarly, we can see the same seating types lead to faster roller coasters.

```python
plt.figure(figsize = (10,10))
plt.bar(x=mean_grouped.seating_type, height=mean_grouped.length)
plt.xticks(rotation = 45, size = 11)
plt.yticks(size = 12)
plt.ylabel('Roller Coaster Length', size = 12)
plt.title('Bar Plot of Roller Coaster Length by Seating Type', size = 16)
plt.show()
```

Wing and Floorless seating types lead to the longest roller coasters.

In general, there are specific seating types associated with higher, longer, and faster roller coasters.

#### Do roller coaster manufacturers have any specialties (focus on speed, height, seating type, or inversions)?

```python
roller_coasters.manufacturer.unique()
```

There is a large number of different manufacturers in the dataset. We can group the dataset by the manufacturer names and summarize by speed, height, seating type, and number of inversions.

```python
characteristics_by_manuf = roller_coasters.groupby('manufacturer')[['speed','height','num_inversions']].mean().reset_index()
```

```python
print(characteristics_by_manuf.to_string())
```

We can see that some manufacturers such as Hensel Phelps Construction Co., International Amusement Devices, and Holzbau Cordes focus alot on speed and less on height. These manufacturers' roller coasters also do not have any inversions. On the other hand, Hebei Zhongye focus much more on inversions. 

```python
roller_coasters.seating_type.unique()
```

```python
seating_grouped_by_manuf = roller_coasters.groupby(['manufacturer','seating_type']).size().reset_index(name = 'num_manuf')
seating_grouped_by_manuf = seating_grouped_by_manuf[(seating_grouped_by_manuf.manufacturer != 'na') 
                                                    & (seating_grouped_by_manuf.seating_type != 'na')]
print(seating_grouped_by_manuf.to_string())
```

From the table above, we can see that different manufacturers only produce rollercoasters with specific seating types which are considered their specialty. For instance, Dinn Corp. only produces Sit Down roller coasters while EOS rides produces Spinning roller coasters. Golden Horse produces roller coasters with different seating types but the Sit Down type is the most common.

```python
grouped_by_park = roller_coasters.groupby('park')['speed','height','num_inversions'].mean().reset_index()
print(grouped_by_park.to_string())
```

```python
seating_by_park = roller_coasters.groupby(['park','seating_type']).size().reset_index(name = 'num_parks')
print(seating_by_park.to_string())
```

The two tables above show that, similar to manufactuers, different parks specialize in speed, height, and inversions. Also, different parks only have roller coasters with specific seating types.
